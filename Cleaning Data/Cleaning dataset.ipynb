```python
def cleaning_nilai(data):
    mp = {
            "Matematika": "Matematika",
            "Inggris": "Inggris",
            "Indonesia": "Indonesia",
            "Fisika": "Fisika",
            "Kimia": "Kimia",
            "Biologi": "Biologi",
            "Ekonomi": "Ekonomi",
            "Geografi": "Geografi",
            "Sosiologi": "Sosiologi",
            "Sejarah": "Sejarah"
        }

    #Mengambil nilai tengah dari tiap nilai
    for column in data:
        matpel = column.split()
        for kata in matpel:
            if kata == mp["Matematika"] or kata == mp["Inggris"] or kata == mp["Indonesia"] or kata == mp["Fisika"] or kata == mp["Kimia"] or kata == mp["Biologi"] or kata == mp["Ekonomi"] or kata == mp["Geografi"] or kata == mp["Sosiologi"] or kata == mp["Sejarah"]:
                for i in range(len(data)):
                    if data[column][i] == "<70":
                        data[column][i] = 69
                    elif data[column][i] == "71 -75":
                        data[column][i] = 73
                    elif data[column][i] == "76 - 80":
                        data[column][i] = 78
                    elif data[column][i] == "81 - 85":
                        data[column][i] = 83
                    elif data[column][i] == "86 - 90":
                        data[column][i] = 88
                    elif data[column][i] == "91 - 95":
                        data[column][i] = 93
                    elif data[column][i] == "> 95":
                        data[column][i] = 96
                    else:
                        data[column][i] = 0

    return data
```

```python
def replace_zero_w_mean(data, mata_pelajaran):
    jumlah_nilai = 0
    banyak_nilai = 0
    for column in data:
        matpel = column.split()
        for kata in matpel:
            if kata == mata_pelajaran:
                for i in range(len(data)):
                    if data[column][i] != 0:
                        jumlah_nilai += data[column][i]
                        banyak_nilai += 1
    try:
        rata2 = jumlah_nilai/banyak_nilai
    except ZeroDivisionError:
        rata2 = 0

    for column in data:
        matpel = column.split()
        for kata in matpel:
            if kata == mata_pelajaran:
                for i in range(len(data)):
                    if data[column][i] == 0:
                        data[column][i] = rata2
                    
    return data
```

```python
def rata2_mp_semester(data, mata_pelajaran):
    data_rata2 = {}
    
    mapel = []
    for i in range(1,6):
        mapel.append(f"{mata_pelajaran} Semester {i}")
        
    data_drop = data.drop(mapel, axis=1)

    for column in data_drop:
        isi = []
        for i in range(len(data)):
            header = column
            isi.append(data[column][i])
        data_rata2[column] = isi

    mean = []
    for i in range(len(data)):
        jumlah_nilai = 0
        jumlah_nilai += data[f"{mata_pelajaran} Semester 1"][i] + data[f"{mata_pelajaran} Semester 2"][i] + data[f"{mata_pelajaran} Semester 3"][i] + data[f"{mata_pelajaran} Semester 4"][i] + data[f"{mata_pelajaran} Semester 5"][i]
        mean.append(jumlah_nilai/5)

    data_rata2[f"Rata-rata {mata_pelajaran}"] = mean

    data_rata2_frame = pd.DataFrame(data=data_rata2)
    return data_rata2_frame
```

```python
def clean(files):
    cleanned_data = []
    
    for file in files:
        data = open_file(file)

        data = cleaning_nilai(data)

        matpel = ["Matematika", "Inggris", "Indonesia", "Fisika", "Kimia", "Biologi", "Ekonomi", "Geografi", "Sosiologi","Sejarah"]

        for item in matpel:
            data = replace_zero_w_mean(data, item)
            data = rata2_mp_semester(data, item)
            
        drop_index = []

        for i in range(len(data)):
            try:
                float(data["Jurusan Diterima"][i])
            except ValueError:
                pass
            else:
                if math.isnan(float(data["Jurusan Diterima"][i])):
                    drop_index.append(i)
                    
        data = data.drop(drop_index)
        
        cleanned_data.append(data)
        
        result = pd.concat(cleanned_data, ignore_index=True)

    return result
```

```python
datasets = ["Data SNMPTN 2016 IPB by @halokampus.csv",
            "Data SNMPTN 2016 ITB by @halokampus.csv",
            "Data SNMPTN 2016 ITS by @halokampus.csv",
            "Data SNMPTN 2016 UB by @halokampus.csv",
            "Data SNMPTN 2016 UGM by @halokampus.csv",
            "Data SNMPTN 2016 UI by @halokampus.csv",
            "Data SNMPTN 2016 Unair by @halokampus.csv",
            "Data SNMPTN 2016 Undip by @halokampus.csv",
            "Data SNMPTN 2016 Unpad by @halokampus.csv",]

result = clean(datasets)

result
```
