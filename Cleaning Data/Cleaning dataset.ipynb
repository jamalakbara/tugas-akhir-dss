# Cleaning Nilai

Proses cleaning nilai terdiri dari beberapa langkah:

1. Mengambil nilai tengah dari tiap mata pelajaran
2. Mengganti setiap NaN/null data dengan mean per mata pelajaran


```python
def cleaning_nilai(data):

    #Mengambil nilai tengah dari tiap nilai
    data = data.replace("<70", 34.5)
    data = data.replace("71 - 75", 73)
    data = data.replace("76 - 80", 78)
    data = data.replace("81 - 85", 83)
    data = data.replace("86 - 90", 88)
    data = data.replace("91 - 95", 93)
    data = data.replace("> 95", 98)
    data = data.replace("No data", np.NaN)

    return data
```

```python
def replace_zero_w_mean(data, mata_pelajaran):
    
    #mencari rata-rata dari per mata pelajaran
    rata2 = (data[f"{mata_pelajaran} Semester 1"].mean() 
            + data[f"{mata_pelajaran} Semester 2"].mean() 
            + data[f"{mata_pelajaran} Semester 3"].mean()
            + data[f"{mata_pelajaran} Semester 4"].mean()
            + data[f"{mata_pelajaran} Semester 5"].mean())/5
    
    #mengganti NaN value dengan nilai rata-rata
    values = {
        f"{mata_pelajaran} Semester 1": rata2,
        f"{mata_pelajaran} Semester 2": rata2,
        f"{mata_pelajaran} Semester 3": rata2,
        f"{mata_pelajaran} Semester 4": rata2,
        f"{mata_pelajaran} Semester 5": rata2,
    }
    data = data.fillna(value=values)
                    
    return data
```

# Menghitung Rata-Rata Per 5 Semester

Dari nilai-nilai yang telah di-clean kemudian akan dicari rata-rata tiap mata pelajarannya per semester

```python
def rata2_mp_semester(data, mata_pelajaran):
    data_rata2 = {}
    
    #drop column tiap mata pelajaran untuk digantikan dengan rata-rata semester
    mapel = []
    for i in range(1,6):
        mapel.append(f"{mata_pelajaran} Semester {i}")

    data_drop = data.drop(mapel, axis=1)
    
    
    for column in data_drop:
        isi = []
        for i in range(len(data)):
            isi.append(data[column][i])
        data_rata2[column] = isi

    #menghitung rata-rata nilai 5 semester
    mean = []
    for i in range(len(data)):
        jumlah_nilai = 0
        jumlah_nilai += data[f"{mata_pelajaran} Semester 1"][i] + data[f"{mata_pelajaran} Semester 2"][i] + data[f"{mata_pelajaran} Semester 3"][i] + data[f"{mata_pelajaran} Semester 4"][i] + data[f"{mata_pelajaran} Semester 5"][i]
        mean.append(jumlah_nilai/5)

    #memasukkan column rata-rata 5 semester tiap mata pelajaran ke dataframe
    data_rata2[f"Rata-rata {mata_pelajaran}"] = mean

    data_rata2_frame = pd.DataFrame(data=data_rata2)
    
    return data_rata2_frame
```

# Menghapus NaN

Proses ini akan menghapus tiap rows yang memiliki NaN value yang ada pada column Jurusan Diterima

```python
def remove_nan(data):
    drop_index = []

    #menghapus NaN value
    for i in range(len(data)):
        try:
            float(data["Jurusan Diterima"][i])
        except ValueError:
            pass
        else:
            if math.isnan(float(data["Jurusan Diterima"][i])) or int(data["Jurusan Diterima"][i]) == 0:
                drop_index.append(i)

    data = data.drop(drop_index)
        
    return data
```

```python
def clean_jurusan_ptn(data):
    for i in range(len(data)):
        for column in data:
            if column == "Jurusan Diterima":
                data[column][i] = data[column][i].title()
            elif column == "PTN Diterima":
                data[column][i] = data[column][i].upper()
    
    return data
```

```python
def clean(files):
    cleanned_data = []
    
    for file in files:
        data = open_file(file)

        data = cleaning_nilai(data)

        matpel = ["Matematika", "Inggris", "Indonesia", "Fisika", "Kimia", "Biologi", "Ekonomi", "Geografi", "Sosiologi","Sejarah"]

        for item in matpel:
            data = replace_zero_w_mean(data, item)
            data = rata2_mp_semester(data, item)
                    
        data = remove_nan(data)
        
        cleanned_data.append(data)
        
        data_merge = pd.concat(cleanned_data, ignore_index=True)
        
        result = clean_jurusan_ptn(data_merge)

    return result
```

```python
def sort_data(data):
    data_sort = data.sort_values(by=['Jurusan Diterima', 'PTN Diterima'])
    
    data_sort.reset_index(inplace = True)

    data_sort = data_sort.drop(["index"], axis=1)

    return data_sort
```

```python
def get_jurusan_ptn(data):
    temp_jurusan = ""
    temp_ptn = ""
    list_jurusan_ptn = []
    for i in range(len(data)):
        if data["Jurusan Diterima"][i] != temp_jurusan and data["PTN Diterima"][i] != temp_ptn:
            temp_jurusan = data["Jurusan Diterima"][i]
            temp_ptn = data["PTN Diterima"][i]
            list_jurusan_ptn.append((temp_jurusan, temp_ptn))
        else:
            pass

    return list_jurusan_ptn
```

```python
def get_tabel_alternatif(data, list_jurusan_ptn):
    frame = {}
    for j in range(len(list_jurusan_ptn)):
        list_nilai = []
        for column in data:
            jum = 0
            byk = 0
            for i in range(len(data)):
                if data["Jurusan Diterima"][i] == list_jurusan_ptn[j][0] and data["PTN Diterima"][i] == list_jurusan_ptn[j][1]:
    #                 print(data_sort["Jurusan Diterima"][i], data_sort["PTN Diterima"][i])
                    temp_jurusan = data["Jurusan Diterima"][i]
                    temp_ptn = data["PTN Diterima"][i]
                    try:
                        float(data[column][i])
                    except ValueError:
                        break
                    else:
                        jum += data[column][i]
                        byk += 1
                else:
                    pass
            try:
    #             print(column, jum/byk)
                list_nilai.append(jum/byk)
            except ZeroDivisionError:
                continue
        nama_alternatif = f"{temp_jurusan} - {temp_ptn}"
        if "administrasi" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "sosial" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "ekonomi" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "hi" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "hukum" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "komunikasi" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "politik" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "sosiologi" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "fsrd" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "sbm" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "akuntansi" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "manajemen" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "psdk" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "fiskal" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif ("hubungan" and "internasional") in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "kriminologi" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif ("psikologi" and "-" and "ui") in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "bisnis" in temp_jurusan.casefold().split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPS)"
        elif "sastra" in temp_jurusan.casefold().split():
            continue
        elif "bahasa" in temp_jurusan.casefold().split():
            continue
        
        if "(IPS)" not in nama_alternatif.split():
            nama_alternatif = f"{temp_jurusan} - {temp_ptn} (IPA)"
            
        frame[nama_alternatif] = list_nilai

    tabel = pd.DataFrame(frame, index=["Matematika", "Inggris", "Indonesia", "Fisika", "Kimia", "Biologi", "Ekonomi", "Geografi", "Sosiologi", "Sejarah"])

    return tabel
```

```python
datasets = ["Data SNMPTN 2016 IPB by @halokampus.csv",
            "Data SNMPTN 2016 ITB by @halokampus.csv",
            "Data SNMPTN 2016 ITS by @halokampus.csv",
            "Data SNMPTN 2016 UB by @halokampus.csv",
            "Data SNMPTN 2016 UGM by @halokampus.csv",
            "Data SNMPTN 2016 UI by @halokampus.csv",
            "Data SNMPTN 2016 Unair by @halokampus.csv",
            "Data SNMPTN 2016 Undip by @halokampus.csv",
            "Data SNMPTN 2016 Unpad by @halokampus.csv",]

result = clean(datasets)

data_sort = sort_data(result)

list_jurusan_ptn = get_jurusan_ptn(data_sort)

tabel_alternatif = get_tabel_alternatif(data_sort, list_jurusan_ptn)

tabel_alternatif
```

```python
nilai_jams = [83.34, 82, 90, 85.2, 82, 83.4, 0, 0, 0, 0]

tabel_alternatif["Threshold"] = nilai_jams

tabel_alternatif
```

```python
pilihan_jurusan = ["Stei - ITB (IPA)", "Administrasi Fiskal - UI (IPS)", "Akuntansi - UGM (IPS)"]
```

```python
deviasi = {}
dev_val = {}

# for index_name in tabel_alternatif.index:
#     dev_val[index_name] = 0
    
for jurusan in pilihan_jurusan:
    for jurusan_compare in pilihan_jurusan:
        if jurusan == jurusan_compare:
            continue
        else:
            for index_name in tabel_alternatif.index:
                if "(IPA)" in jurusan.split():
                    if index_name == "Matematika" or index_name == "Inggris" or index_name == "Indonesia" or index_name == "Fisika" or index_name == "Kimia" or index_name == "Biologi":
                        
                        #menghitung deviasi
                        nilai_deviasi = tabel_alternatif[jurusan][index_name] - tabel_alternatif[jurusan_compare][index_name]
                        
                        #menghitung fungsi preferensi gaussian
                        if nilai_deviasi > 0:
                            gaussian = 1 - math.exp(-1*(((nilai_deviasi)**2) / (2*((tabel_alternatif["Threshold"][index_name])**2))))
                        else:
                            gaussian = 0
                            
                        deviasi_name = f"{jurusan}&{jurusan_compare}"
                        dev_val[index_name] = gaussian
                    
                elif "(IPS)" in jurusan.split():
                    if index_name == "Ekonomi" or index_name == "Inggris" or index_name == "Indonesia" or index_name == "Geografi" or index_name == "Sosiologi" or index_name == "Sejarah":
                        
                        #menghitung deviasi
                        nilai_deviasi = tabel_alternatif[jurusan][index_name] - tabel_alternatif[jurusan_compare][index_name]
                        
                        #menghitung fungsi preferensi gaussian
                        if nilai_deviasi > 0:
                            gaussian = 1 - math.exp(-1*(((nilai_deviasi)**2) / (2*((tabel_alternatif["Threshold"][index_name])**2))))
                        else:
                            gaussian = 0
                            
                        deviasi_name = f"{jurusan}&{jurusan_compare}"
                        dev_val[index_name] = gaussian
    
                deviasi[deviasi_name] = dev_val
        dev_val = {}
    
deviasi
```

```python
index_preferensi_gaussian = {}
weight = 1

for compare, fgaussian in deviasi.items():
    nilai_gaussian = 0
    for kriteria in fgaussian.keys():
        nilai_gaussian += fgaussian[kriteria]*weight
    nama_gaussian = compare
    index_preferensi_gaussian[nama_gaussian] = nilai_gaussian
    
index_preferensi_gaussian
```

```python
entering_flow = {}

for item in pilihan_jurusan:
    ef = 0
    for piece in index_preferensi_gaussian.keys():
        s = piece.split("&")
        
        if item in s and s.index(item) == 1:
            ef += index_preferensi_gaussian[piece] * (1/(len(pilihan_jurusan) - 1))
            
    entering_flow[item] = ef
    
entering_flow
```

```python
leaving_flow = {}

for item in pilihan_jurusan:
    lf = 0
    for piece in index_preferensi_gaussian.keys():
        s = piece.split("&")
        
        if item in s and s.index(item) == 0:
            lf += index_preferensi_gaussian[piece] * (1/(len(pilihan_jurusan) - 1))
            
    leaving_flow[item] = lf
    
leaving_flow
```

```python
net_flow = {}

for jurusan in pilihan_jurusan:
    net_flow[jurusan] = entering_flow[jurusan] - leaving_flow[jurusan]
    
net_flow
```
