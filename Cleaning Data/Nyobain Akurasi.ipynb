```python
import pandas as pd
import math
```

```python
def ipa_ips(jurusan, kampus):
    
    if "administrasi" in jurusan.casefold().split()\
    or "sosial" in jurusan.casefold().split()\
    or "ekonomi" in jurusan.casefold().split()\
    or "hi" in jurusan.casefold().split()\
    or "hukum" in jurusan.casefold().split()\
    or "komunikasi" in jurusan.casefold().split()\
    or "politik" in jurusan.casefold().split()\
    or "sosiologi" in jurusan.casefold().split()\
    or "fsrd" in jurusan.casefold().split()\
    or "sbm" in jurusan.casefold().split()\
    or "akuntansi" in jurusan.casefold().split()\
    or "manajemen" in jurusan.casefold().split()\
    or "psdk" in jurusan.casefold().split()\
    or "fiskal" in jurusan.casefold().split()\
    or ("hubungan" in jurusan.casefold().split() and "internasional" in jurusan.casefold().split())\
    or "kriminologi" in jurusan.casefold().split()\
    or "psikologi" in jurusan.casefold().split()\
    or "bisnis" in jurusan.casefold().split()\
    or "filsafat" in jurusan.casefold().split()\
    or "jurnalistik" in jurusan.casefold().split()\
    or "humas" in jurusan.casefold().split()\
    or "desain" in jurusan.casefold().split()\
    or "film" in jurusan.casefold().split()\
    or ("hubungan" in jurusan.casefold().split() and "masyarakat" in jurusan.casefold().split()):
        s = f"{jurusan} - {kampus} (IPS)"
        
    elif "sastra" in jurusan.casefold().split()\
    or "bahasa" in jurusan.casefold().split()\
    or "perpustakaan" in jurusan.casefold().split():
        s = f"{jurusan} - {kampus} (Bahasa)"
    elif "ugm" in jurusan.casefold().split():
        s = "error"

    else:
        s = f"{jurusan} - {kampus} (IPA)"
        
    return s
```

```python
nilai_test = pd.read_csv("Splitted Test.csv", index_col = 0)

nilai_test
```

```python
for column in nilai_test:
    if "rata-rata" not in column.casefold().split():
        nilai_test[column] = nilai_test[column].astype('str')
    else:
        nilai_test[column] = nilai_test[column].astype('float64')
```

```python
daftar_nilai = {}

for index in nilai_test.index:
    nilai = []
    for column in nilai_test:
        if "rata-rata" in column.casefold().split():
            if type(nilai_test["Jurusan Sekolah"][index]) == str and nilai_test["Jurusan Sekolah"][index] == "IPS":
                if "fisika" in column.casefold().split()\
                or "kimia" in column.casefold().split()\
                or "biologi" in column.casefold().split():
                    nilai.append(0)
                else:
                    nilai.append(nilai_test[column][index])
            if type(nilai_test["Jurusan Sekolah"][index]) == str and nilai_test["Jurusan Sekolah"][index] == "IPA":
                if "ekonomi" in column.casefold().split()\
                or "geografi" in column.casefold().split()\
                or "sosiologi" in column.casefold().split():
                    nilai.append(0)
                else:
                    nilai.append(nilai_test[column][index])
    if type(nilai_test["Nama Panggilan"][index]) == str:
        daftar_nilai[nilai_test["Nama Panggilan"][index]] = nilai
```

```python
daftar_pilihan_jurusan = {}

for index in nilai_test.index:
    jurusan = []
    if type(nilai_test["PTN Pilihan I"][index]) == str\
    and nilai_test["PTN Pilihan I"][index].casefold() != "nan"\
    and nilai_test["PTN Pilihan I"][index].casefold() != "kosong"\
    and nilai_test["PTN Pilihan I"][index].casefold() != "-":
        if type(nilai_test["Jurusan Pilihan I di PTN Pilihan I"][index]) == str\
        and nilai_test["Jurusan Pilihan I di PTN Pilihan I"][index].casefold() != "nan"\
        and nilai_test["Jurusan Pilihan I di PTN Pilihan I"][index].casefold() != "kosong"\
        and nilai_test["Jurusan Pilihan I di PTN Pilihan I"][index].casefold() != "-":
            s = ipa_ips(nilai_test["Jurusan Pilihan I di PTN Pilihan I"][index], nilai_test["PTN Pilihan I"][index])
            jurusan.append(s)
            
        if type(nilai_test["Jurusan Pilihan II di PTN Pilihan I"][index]) == str\
        and nilai_test["Jurusan Pilihan II di PTN Pilihan I"][index].casefold() != "nan"\
        and nilai_test["Jurusan Pilihan II di PTN Pilihan I"][index].casefold() != "kosong"\
        and nilai_test["Jurusan Pilihan II di PTN Pilihan I"][index].casefold() != "-":
            s = ipa_ips(nilai_test["Jurusan Pilihan II di PTN Pilihan I"][index], nilai_test["PTN Pilihan I"][index])
            jurusan.append(s)
            
    if type(nilai_test["PTN Pilihan II"][index]) == str\
    and nilai_test["PTN Pilihan II"][index].casefold() != "nan"\
    and nilai_test["PTN Pilihan II"][index].casefold() != "kosong"\
    and nilai_test["PTN Pilihan II"][index].casefold() != "-":
        if type(nilai_test["Jurusan Pilihan I di PTN Pilihan II"][index]) == str\
        and nilai_test["Jurusan Pilihan I di PTN Pilihan II"][index].casefold() != "nan"\
        and nilai_test["Jurusan Pilihan I di PTN Pilihan II"][index].casefold() != "kosong"\
        and nilai_test["Jurusan Pilihan I di PTN Pilihan II"][index].casefold() != "-":
            s = ipa_ips(nilai_test["Jurusan Pilihan I di PTN Pilihan II"][index], nilai_test["PTN Pilihan II"][index])
            jurusan.append(s)
            
        if type(nilai_test["Jurusan Pilihan II di PTN Pilihan II"][index]) == str\
        and nilai_test["Jurusan Pilihan II di PTN Pilihan II"][index].casefold() != "nan"\
        and nilai_test["Jurusan Pilihan II di PTN Pilihan II"][index].casefold() != "kosong"\
        and nilai_test["Jurusan Pilihan II di PTN Pilihan II"][index].casefold() != "-":
            s = ipa_ips(nilai_test["Jurusan Pilihan II di PTN Pilihan II"][index], nilai_test["PTN Pilihan II"][index])
            jurusan.append(s)
            
    if type(nilai_test["Nama Panggilan"][index]) == str:
        daftar_pilihan_jurusan[nilai_test["Nama Panggilan"][index]] = jurusan
        
daftar_pilihan_jurusan
```
